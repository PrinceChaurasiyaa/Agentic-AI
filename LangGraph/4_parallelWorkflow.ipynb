```python
# define State
class BatsmanState(TypedDict):

    runs: int
    balls: int
    fours: int
    sixes: int

    sr: float
    bpb: float
    boundary_percent: float
    summary: str
```

```python
# calcualte Strike Rate
def calculate_sr(state: BatsmanState) -> BatsmanState:
    sr = state['balls'] / (state['fours'] + state['sixes'])
    return {'sr': sr}
```
